```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
def phi(x, y, eps=1):
    t1 = np.exp(-eps**2 * (x-y)**2, dtype=np.longdouble)
    t2 = np.exp(-eps**2 * (x**2 + y**2), dtype=np.longdouble)
    denum = 1 - np.exp(-2*eps**2, dtype=np.longdouble)
    num_p1 = np.exp(-eps**2 * (1-y)**2, dtype=np.longdouble) - np.exp(-eps**2 * (1+y**2), dtype=np.longdouble)
    num_p2 = np.exp(-eps**2 * (x-1)**2, dtype=np.longdouble) - np.exp(-eps**2 * (1+x**2), dtype=np.longdouble)
    num = num_p1 * num_p2
    t3 = num / denum
    return t1 - t2 - t3
```

```python
def phi_x(x, y, eps=1):
    t1 = -2 * eps**2 * (x-y) * np.exp(-eps**2 * (x-y)**2, dtype=np.longdouble)
    t2 = 2 * eps**2 * x * np.exp(-eps**2 * (y**2 + x**2), dtype=np.longdouble)
    denum = 1 - np.exp(-2*eps**2, dtype=np.longdouble)
    num_p1 = np.exp(-eps**2 * (1-y)**2, dtype=np.longdouble) - np.exp(-eps**2 * (1+y**2), dtype=np.longdouble)
    num_p2 = -2 * eps**2 * (x-1) * np.exp(-eps**2 * (x-1)**2) + 2 * eps**2 * x * np.exp(-eps**2 * (1+x**2), dtype=np.longdouble)
    num = num_p1 * num_p2
    t3 = num / denum
    return t1 + t2 - t3
```

```python
def phi_xx(x, y, eps=1):
    t1 = -2 * eps**2 * np.exp(-eps**2 * (x-y)**2, dtype=np.longdouble)
    t2 = 4 * eps**4 * (x-y)**2 * np.exp(-eps**2 * (x-y)**2, dtype=np.longdouble)
    t3 = 2 * eps**2 * np.exp(-eps**2 * (x**2 + y**2))
    t4 = -4 * eps**4 * x**2 * np.exp(-eps**2 * (x**2 + y**2))
    
    denum = 1 - np.exp(-2*eps**2)
    num_p1 = np.exp(-eps**2 * (1-y)**2) - np.exp(-eps**2 * (1 + y**2))
    num_p2_t1 = -2 * eps**2 * np.exp(-eps**2 * (x-1)**2, dtype=np.longdouble)
    num_p2_t2 = 4 * eps**4 * (x-1)**2 * np.exp(-eps**2 * (x - 1)**2, dtype=np.longdouble)
    num_p2_t3 = 2 * eps**2 * np.exp(-eps**2 * (1+x**2), dtype=np.longdouble)
    num_p2_t4 = -4 * eps**4 * x**2 * np.exp(-eps**2 * (1 + x**2), dtype=np.longdouble)
    num_p2 = num_p2_t1 + num_p2_t2 + num_p2_t3 + num_p2_t4
    num = num_p1 * num_p2
    t5 = num / denum
    return t1 + t2 + t3 + t4 - t5
```

```python
y = np.linspace(0, 1, 11, dtype=np.longdouble)
x = np.linspace(0, 1, 101, dtype=np.longdouble)
```

```python
phi(0, y)
```

```python
phi(1, y)
```

```python
plt.figure(figsize=(8, 6))
eps = 1
plt.plot(x, phi(x, y[1], eps=eps), label='{:.1f}'.format(y[1]));
plt.plot(x, phi(x, y[2], eps=eps), label='{:.1f}'.format(y[2]));
plt.plot(x, phi(x, y[3], eps=eps), label='{:.1f}'.format(y[3]));
plt.plot(x, phi(x, y[4], eps=eps), label='{:.1f}'.format(y[4]));
plt.plot(x, phi(x, y[5], eps=eps), label='{:.1f}'.format(y[5]));
plt.plot(x, phi(x, y[6], eps=eps), label='{:.1f}'.format(y[6]));
plt.plot(x, phi(x, y[7], eps=eps), label='{:.1f}'.format(y[7]));
plt.plot(x, phi(x, y[8], eps=eps), label='{:.1f}'.format(y[8]));
plt.plot(x, phi(x, y[9], eps=eps), label='{:.1f}'.format(y[9]));

plt.legend();
```

```python
plt.figure(figsize=(8, 6))
# eps = 6
plt.plot(x, phi_x(x, y[1], eps=eps), label='{:.1f}'.format(y[1]));
plt.plot(x, phi_x(x, y[2], eps=eps), label='{:.1f}'.format(y[2]));
plt.plot(x, phi_x(x, y[3], eps=eps), label='{:.1f}'.format(y[3]));
plt.plot(x, phi_x(x, y[4], eps=eps), label='{:.1f}'.format(y[4]));
plt.plot(x, phi_x(x, y[5], eps=eps), label='{:.1f}'.format(y[5]));
plt.plot(x, phi_x(x, y[6], eps=eps), label='{:.1f}'.format(y[6]));
plt.plot(x, phi_x(x, y[7], eps=eps), label='{:.1f}'.format(y[7]));
plt.plot(x, phi_x(x, y[8], eps=eps), label='{:.1f}'.format(y[8]));
plt.plot(x, phi_x(x, y[9], eps=eps), label='{:.1f}'.format(y[9]));

plt.legend();
```

```python
plt.figure(figsize=(8, 6))
# eps = 6
plt.plot(x, phi_xx(x, y[1], eps=eps), label='{:.1f}'.format(y[1]));
plt.plot(x, phi_xx(x, y[2], eps=eps), label='{:.1f}'.format(y[2]));
plt.plot(x, phi_xx(x, y[3], eps=eps), label='{:.1f}'.format(y[3]));
plt.plot(x, phi_xx(x, y[4], eps=eps), label='{:.1f}'.format(y[4]));
plt.plot(x, phi_xx(x, y[5], eps=eps), label='{:.1f}'.format(y[5]));
plt.plot(x, phi_xx(x, y[6], eps=eps), label='{:.1f}'.format(y[6]));
plt.plot(x, phi_xx(x, y[7], eps=eps), label='{:.1f}'.format(y[7]));
plt.plot(x, phi_xx(x, y[8], eps=eps), label='{:.1f}'.format(y[8]));
plt.plot(x, phi_xx(x, y[9], eps=eps), label='{:.1f}'.format(y[9]));

plt.legend();
```
